```python
import numpy as np
import pandas as  pd
```

```python
df=pd.read_csv('/content/dsjob - dsjob.csv')
df.head(2)
```

```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.5 and df[var].isnull().mean()>0]
cols
```

```python
new_df=df[cols].dropna()
df.shape,new_df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
#orignal data
df['experience'].hist(bins=50,ax=ax,density=True,color='red')

#data after cca,the argument alpha makes the color transparent , so can
# see the overlay of the 2 distributions
new_df['experience'].hist(bins=50,ax=ax,density=True,color='blue',alpha=0.8)
```
