```python
from textblob import TextBlob    #textblob ko acess krna

def textblob_sentiment_analysis(review):
    sentiment = TextBlob(review).sentiment

    if sentiment.polarity > 0.1:
        return "positive"
    elif sentiment.polarity < -0.1:
        return "negative"
    else:
        return "neutral"
```

```python
df['Sentiment']=df['Review'].apply(textblob_sentiment_analysis) #apply using like map fuction and perform the all data
```

```python
sentiment_distribution=df['Sentiment'].value_counts()
sentiment_distribution                                 #value count in spacific data
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=sentiment_distribution.index,y=sentiment_distribution.values) #indexing btana
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df,x='Rating',
hue='Sentiment') #column k base pr dusre column ko compair krna ....
plt.xlabel('Rating')
plt.ylabel('count')
plt.legend(title='Sentiment')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (Tum apna real df use kar lena)
df = pd.DataFrame({
    'Rating': [4.5, 4.2, 3.8, 4.7, 3.5, 4.9, 2.8],
    'Reviews': [500, 1500, 3000, 8000, 200, 12000, 4500]
})

# 1ï¸âƒ£ Missing values handle (just in case)
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
df['Reviews'] = df['Reviews'].fillna(df['Reviews'].mean())

# 2ï¸âƒ£ Review Category create (LOW, MEDIUM, HIGH)
df['Review_Category'] = pd.cut(
    df['Reviews'],
    bins=[0, 1000, 5000, df['Reviews'].max()+1],
    labels=['Low', 'Medium', 'High']
)

# 3ï¸âƒ£ Group By Rating mean
grouped_data = df.groupby('Review_Category')['Rating'].mean()
print("Group By Result:\n", grouped_data)

# 4ï¸âƒ£ Boxplot
sns.boxplot(x='Review_Category', y='Rating', data=df)
plt.title('Rating Distribution by Review Category')
plt.xlabel('Review Category')
plt.ylabel('Rating')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Sales": [120, 150, 170, 160, 180, 190, 200]
})

fig = px.line(data, x="Day", y="Sales", title="Weekly Sales (Line Chart)", markers=True)
fig.update_layout(xaxis_title="Day of Week", yaxis_title="Sales (â‚¹)")
fig.show()
```

```python
fig = px.bar(data, x="Day", y="Sales", title="Weekly Sales (Bar Chart)")
fig.update_traces(texttemplate='%{y}', textposition='outside')  # bar ke upar values
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Pie(labels=data['Day'], values=data['Sales'], hole=0.3)])
fig.update_layout(title_text="Sales Distribution by Day (Pie Chart)")
fig.show()
```

```python
df = pd.DataFrame({
    "Day": ["Mon","Tue","Wed","Thu","Fri"],
    "Store_A": [100,120,130,140,150],
    "Store_B": [90,110,125,135,160]
})

fig = go.Figure()
fig.add_trace(go.Bar(x=df['Day'], y=df['Store_A'], name='Store A'))
fig.add_trace(go.Bar(x=df['Day'], y=df['Store_B'], name='Store B'))
fig.update_layout(barmode='group', title="Sales: Store A vs Store B")
fig.show()
```

```python
fig = px.scatter(df, x='Store_A', y='Store_B', text=df['Day'],
                 title="Store A vs Store B (Scatter)")
fig.update_traces(marker=dict(size=12, line=dict(width=1)))
fig.update_layout(xaxis_title="Store A Sales", yaxis_title="Store B Sales")
fig.show()









```
