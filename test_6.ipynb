```python
x=df.drop(columns=['Loan Amount'])
y=df['Loan Amount']
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
log_transformer=FunctionTransformer(np.log1p)
```

```python
from sklearn.preprocessing import LabelEncoder

x_encoded = x.copy()
for column in ['Gender', 'Marital Status', 'Education Level','Employment Status','Type of Loan']:
    lb = LabelEncoder()
    x_encoded[column] = lb.fit_transform(x_encoded[column])

x_transformed=log_transformer.transform(x_encoded)
```
