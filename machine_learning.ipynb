```python
import pandas as pd
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df=df.drop(columns=["Last_Name","Address","Not_Useful_Column"])
```

```python
df.head(5)
```

```python
df['First_Name']=df['First_Name'].map({'Frodo':0,"Abed":1,"Walter":2,"Dwight":3,"Jon":4})
df['Do_Not_Contact']=df['Do_Not_Contact'].map({'No':0,"Yes":1,"NaN":2,"Y":3,"No":4})
df['Paying Customer']=df['Paying Customer'].map({'No':0,"Yes":1,"NaN":2,"Y":3,"No":4})
```

```python
df.head(5)
```

```python
x=df.drop(columns=["Paying Customer"])
y=df["Paying Customer"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
